```python
import pandas as pd
```

```python
filename = "bestsellers with categories.csv"
file = pd.read_csv(filename)
```

```python
"""Виведіть перші п'ять рядків"""
file.head(5)
```

```python
"""Виведіть розміри датасету"""
rows, cols = file.shape
print(f"rows: {rows}, cols: {cols}")
```

```python
"""Про скільки книг зберігає дані датасет?"""
# 550
```

```python
file.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
```

```python
"""Чи є в якихось змінних пропуски? (Так / ні)"""
file.isnull().sum()
# Ні, немає
```

```python
"""Перевірте, які є унікальні значення в колонці genre (використовуйте функцію unique)"""
a = pd.unique(file["genre"]) 
# Non Fiction та Fiction
print("Унікальні значення: ", end='')
print(*a, sep=', ')
```

```python
"""Визначте, яка ціна у нас максимальна, мінімальна, середня, медіанна (використовуйте функції max, min, mean, median)"""
prices = file["price"]
max_price = prices.max()
min_price = prices.min()
mean_price = prices.mean()
median_price = prices.median()
print(f"Максимальна ціна: {max_price}\nМінімальна ціна: {min_price}\nСередня ціна:{mean_price}\nМедіанна ціна: {median_price}")
```

```python
"""
Який рейтинг у датасеті найвищий?
Скільки книг мають такий рейтинг?
Яка книга має найбільше відгуків?
З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)?
Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)?
Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)?
І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).
Яка книга остання у відсортованому списку?"""

max_rate = file["user_rating"].max()
books_with_max_rate = file[file["user_rating"] == max_rate].name.drop_duplicates().count()
book_with_max_reviews = file[file["reviews"] == file["reviews"].max()]["name"]
max_price_2015 = file[file["year"] == 2015].price.max()
max_price_book_2015 = file[(file["year"] == 2015) & (file["price"] == max_price_2015)].name
genre_fiction_books_amount = file[(file["year"] == 2010) & (file["genre"] == "Fiction")]
books_rate_49_2010_2011 = file[(file["year"].isin([2010, 2011])) & (file["user_rating"] == 4.9)]
sorted_price_2015_ASC = file[(file["year"] == 2015) & (file["price"] < 8)].sort_values(by='price')
last_book_in_sorted = sorted_price_2015_ASC.tail(1).name
print(f"Найвищий рейтинг: {max_rate}\nКількість книг з найбільшим рейтингом: {books_with_max_rate}\nКнига з найбільшою кількістю відгудків: {book_with_max_reviews}\nНайдорожча книга у 2015: {max_price_book_2015}\nКількість книг з жанру Fiction у 2015: {genre_fiction_books_amount}\nКількість книг з рейтингом 4.9 у 2010 та 2011 роках: {books_rate_49_2010_2011}\nОстання книга у відсортованому списку: {last_book_in_sorted}")
```

```python
grouped_by_genre = file.groupby('genre').agg({'price': ['max', 'min']})
grouped_by_genre
# Fiction max price = 82
# Fiction min price = 0
# Non Fiction max price = 0
# Non Fiction min price = 105
```

```python
books_by_author = file.groupby('author').agg({'name': ['count']})
books_by_author.sort_values(('name', 'count'), ascending=False)
# 248 x 1
# Jeff Kiney
# Books 12 
```

```python
authors_by_rate = file.groupby('author').agg({'user_rating': ['mean']}).round(2)
authors_by_rate.sort_values(('user_rating', 'mean'), ascending=False)
# Min rate Donna Tartt
# Donna Tartt rate 3.9
```

```python
authors_by_rate_and_books = pd.concat([books_by_author, authors_by_rate], axis=1)
authors_by_rate_and_books.sort_values(([('name','count'), ('user_rating', 'mean')]), ascending=[True, True]).head(1)
# Muriel Barbery
```
