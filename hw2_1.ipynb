```python
import matplotlib.pyplot as plot
import numpy as np
import pandas as pd
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
table_name = "Коефіцієнт народжуваності в регіонах України"
table = pd.read_html(url, thousands=',', decimal='.', match=table_name)[0]
table
```

```python
"""Вивести перші рядки таблиці за допомогою методу head"""
table.head()
```

```python
"""Визначте кількість рядків та стовпців у датафреймі (атрибут shape)"""
rows, cols = table.shape
print(f"rows: {rows}, cols: {cols}")
```

```python
"""Замініть у таблиці значення "—" на значення NaN"""
table.replace("—", np.nan, inplace=True)
table
```

```python
"""Визначте типи всіх стовпців за допомогою dataframe.dtypes"""
table.dtypes
```

```python
"""Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ """
for col in table:
    if col.isdigit():
        table[col] = table[col].astype(float)
table.dtypes
```

```python
"""Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)"""
(table.isnull().sum()/rows * 100).round(2)
```

```python
"""Видаліть з таблиці дані по всій країні, останній рядок таблиці"""
table.drop(table.index[-1])
```

```python
"""Замініть відсутні дані в стовпцях середніми значеннями цих стовпців"""
table.fillna(table.mean(numeric_only=True).round(2))
```

```python
"""Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні"""
table[table['2019'] > table['2019'].mean()][['Регіон','2019']]
```

```python
"""У якому регіоні була найвища народжуваність у 2014 році?"""
table.loc[table['2014'].idxmax()]['Регіон']
```

```python
"""Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році"""

plot.bar(table['Регіон'], table['2019'])
plot.tick_params(labelrotation=90)
plot.title('Діаграма народжуваності по регіонах України у 2019 році')
plot.ylabel('Народжуваність')
plot.show()
```
